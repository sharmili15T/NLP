### Regular Expression

```python
import re
```

```python
s1= 'Hello there!! Python Tutorial'
s2= 'Get ready to learn Python'
```

```python
pattern = 'python'
re.match(pattern,s1)#no output since case sensitive
```

```python
#to ignore the case we have a parameter

m= re.match(pattern,s1, flags=re.IGNORECASE)
print(m)
```

```python
m= re.match(pattern,s1)
print('found{} at {} and {} in the string {}'.format(
m.group(0),m.start(),m.end(),s1
))
```

```python
f= open('myfile.txt')
for line in f:
    line= line.rstrip()
    if re.search('how',line):
        print(line)

f.close()
```

```python
s= 'A message from cse@umich.edu to wyv@ude.edu about meeting @2pm'

ls = re.findall('\S+@\S+', s)
print(ls)
ls
```
